```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from unity_environment_wrapper import unity_env
from DDPG_agent import DDPG_Agent
```

```python
env = unity_env(file_name="Reacher_Windows_x86_64/Reacher.exe", no_graphics=True, verbose_level=1)

agent = DDPG_Agent(state_size = env.state_size, action_size = env.action_size,\
                   random_seed=1,\
                   replay_buffer_size = int(1e6),\
                   replay_batch_size =128,\
                   update_every=20,\
                   update_times=4,\
                   lr_actor=1e-4,\
                   lr_critic=1e-3,\
                   noise_sigma=0.2,\
                   noise_theta=0.15,\
                   no_reward_dropout  = 0.9,\
                   no_reward_value = -0.1,\
                   actor_arch = ['b',128,'b','r',256,'b','r'],\
                   critic_arch = [['b',128],['b',64],['b','r',256,'b','r']],\
                   gamma = 0.95,\
                   use_cuda=True, verbose_level=2)

output_name = "results"
```

```python
all_scores, best_test_score, best_test_average, best_test_stdev = env.train(agent,output_name,num_episode_search=500)
```

## plot the scores

```python
fig = plt.figure()
ax  = fig.add_subplot(111)
plt.plot(np.arange(len(all_scores)), all_scores)
plt.ylabel('Score')
plt.xlabel('Train Episode #')
plt.show()
```
